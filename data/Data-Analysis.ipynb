##Automatically generate a professional Word document (.docx)

```python
pip install python-docx
```

```python
import polars as pl
from docx import Document
from datetime import datetime
```

```python
df = pl.read_csv("insurance.csv")
```

###What Your Word Document Must Contain

Project Title

Dataset Description

Number of Rows & Columns

Column Names & Data Types

Missing Values Summary

Basic Statistics

Date of Report

```python
document = Document()
document.add_heading("Data Documentation Report", level=1)
document.add_paragraph(f"Report Generated On: {datetime.now()}")
document.add_heading("1. Dataset Overview", level=2)
document.add_paragraph(f"Number of Rows: {df.shape[0]}")
document.add_paragraph(f"Number of Columns: {df.shape[1]}")
document.add_heading("2. Column Details", level=2)

table = document.add_table(rows=1, cols=2)
table.rows[0].cells[0].text = "Column Name"
table.rows[0].cells[1].text = "Data Type"
for col, dtype in df.schema.items():
    row = table.add_row().cells
    row[0].text = col
    row[1].text = str(dtype)
document.add_heading("3. Missing Values Summary", level=2)

null_counts = df.null_count()
for col in df.columns:
    document.add_paragraph(f"{col}: {null_counts[col][0]} missing values")
document.add_heading("4. Basic Statistics", level=2)

stats = df.describe().to_pandas()

stats_table = document.add_table(rows=stats.shape[0]+1, cols=stats.shape[1
                                                                         ])
for j, column in enumerate(stats.columns):
    stats_table.rows[0].cells[j].text = column

# Add data
for i in range(stats.shape[0]):
    for j in range(stats.shape[1]):
        stats_table.rows[i+1].cells[j].text = str(stats.iloc[i, j])
        document.save("Data_Documentation_Report.docx")

print("Word document generated successfully!")


```

--->After generating the Word file using

```python
document.save("Data_Documentation_Report.docx")
```
